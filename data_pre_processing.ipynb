# **PART 2.A: DROPPING THE ENTIRE ROW THAT HAS MISSING VALUE**
There are two ways to handle missing values. To train the model, the missing values and the values associated with them are not uswful at all.

### In the first case, ***when the number of missing values is way less than the total dataset***, we will simply delete the entire row that contain the missing value.For example, our current dataset has only 3 missing values. So, deleting those rows will be the easiest way.

### In the second case, ***when the number of missing values is significant compared to entire dataset***, deleting those rows would be inappropriate, as we will be missing a lot of data. In that scenario, the better way is to input some values in the missing places.

```python
df1.dropna(subset=['Age'], inplace=True)     # here, 'dropna' means dropping the mssing values of'Age'(which is a subset of entire dataset)
                                              # dropping those values for whom null is true.
df1.dropna(subset=['Gender'], inplace=True)
df1.dropna(subset=['weakness'], inplace=True)
```

# **PART 4. IMBALANCE DATASET ISSUE**

For a target to be balanced, the unique targets should have equal proportion. For example, in a dataset of 1000 patients, if none has the disease, the dataset is hugely imbalanced. Even if 20 of them has disease, dataset is still imbalanced.

But if 300/400 of them has the disease, and others have not, then the dataset is balanced.

```python
df['class'].unique() # finds out how many unique values there are in the mentioned column. As all the data are encoded, each unique value is assigned a number
```

```python
df['class'].nunique() # counts how many unique value in that column
```

```python
value_counts=df.groupby('class').size().reset_index(name='count')
print(value_counts)   # prints the number of each unique value under 'class'
```

There are two ways to solve imbalance issue. **Undersampling and Oversampling.**

# **4.A : Undersampling**

Undersampling means reducing the major class samples to match the minor class samples.

For example, there are 200 values of zero, and 317 values of one above. So, the number of one (which is the major class sample here), will be reduced to 200, to match the minor class sample.

```python
from sklearn.utils import resample # importing "resample" module

majority_class=df[df['class']==1]
minority_class= df[df['class']==0]

n_samples=len(minority_class) # n_samples=200

majority_downsampled = resample(majority_class, replace=False, n_samples=n_samples, random_state=42) # majority class is downsampled to 200
balanced_df=pd.concat([minority_class, majority_downsampled])
```

```python
value_counts = balanced_df.groupby('class').size().reset_index(name='count')
print(value_counts)  # prints the number of each unique value under 'class', where the major class is now downsampled to 200.
```

# **3B. OverSampling (SMOTE)**

Increasing the sample of minor classes


```python
from imblearn.over_sampling import SMOTE
```

```python
x = df.drop('class', axis=1) # dropping the target column, here 'class'. Hence only the features remain. X denotes feature in ML.
y= df['class'] # Y denotes target in ML. as our target is "class", it is assigned to y.

smote = SMOTE(random_state=42)  # will be explained later

x_resampled, y_resampled = smote.fit_resample(x,y) # here, "fit" almost works as training the model. here it resamples "smote"

oversampled_df = pd.DataFrame(x_resampled, columns=x.columns) # WILL BE EXPLAINED LATER

oversampled_df['class'] = y_resampled
```

```python
value_counts = oversampled_df.groupby('class').size().reset_index(name='count')
print(value_counts)             # TILL THIS LECTURE 04
```

**5.C: Z-SCORE APPROACH [-3 TO 3]**

z-score for a value= (value-mean)/ standard deviation.

the value of s-score is bound to -3 to +3. if for any value the z-score is beyond this range, that value is OUTLIER.

```python
z_score = (df['Age']-df['Age'].mean())/df['Age'].std()
```

This z_score is an array.

```python
print(z_score)
```

Now, checking existence of OUTLIER in the values of z-score using loop

```python
for i in z_score:
  if i<-3 or i>3: # z-score value is acceptable from -3 to +3
    print("you have outlier", i)
  else:
    continue
```

Now to find out the location of the outlier. The index value of the outlier and the index value of that outlier's z-score is same. So, we need to find out the index of that z-score.

```python
index= z_score.index[z_score == 20.289079482050578][0]
```

It's on user's end whether to remove duplicate value. But it is necessary, to prevent DATA LEAKAGE.

If there are duplicate values in the dataset, it might happen that same value may appear in both train stage and test stage. In that case, due to familiarity of the data, the model might display false accuracy, without proper training.

There's also a negative side of removing the duplicate data. Each model has a mathematical distribution (Gaussian, exponential etc.). Removing duplicate value might alter that distribution.

# **PART 7: FEATURE ENGINEERING**

Feature engineering is the process of creating new features (variables) or transforming existing features to improve the performance of machine learning models.

In machine learning, correlation refers to the degree to which two variables change together.

The correlation coefficient, often denoted as "r," ranges from -1 to 1. A correlation of 1 indicates a perfect positive correlation, meaning that as one variable increases, the other variable also increases. A correlation of -1 indicates a perfect negative correlation, meaning that as one variable increases, the other variable decreases. A correlation of 0 indicates no linear relationship between the variables.

Highly correlated features provide redundant (repetitive in nature) information and may negatively impact the performance of machine learning models.


**Highly correlated features** are often **removed** during feature engineering for several reasons:

**Redundancy:**  Highly correlated features contain similar information, leading to redundancy in the model.

**Multicollinearity:**  High correlation between features can cause multicollinearity, which destabilizes model estimates and reduces interpretability.

**Computational Efficiency:**  Removing correlated features reduces the model's complexity, making it more computationally efficient, especially for large datasets.



But in this scenario, we will consider from a different angle. Our target is "class". The features who are highly correlated with "class", affect "class" the most, unlike those features who are low correlated with our target. Hence, if we only work with highly correlated features (with "class", both postive and negative) and omit the remaining features, the model will be less complex.
